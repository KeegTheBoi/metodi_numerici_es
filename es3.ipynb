**Esercizio 1**

Confrontare il tempo di esecuzione dei due metodi per la creazione di un numpy array di 2000000 di elementi equidistanti nell'intervallo [-4,4]:

    1) costruzione con utilizzo di np.linspace
    
    2) Assegnazione manuale dei valori nel ciclo.
    

```python
def np_array(n):
    return np.linspace(-4,4,n)

def simple_array(n):
    return [x * 8 / n - 4 for x in range(0,n)]
```

```python
import matplotlib.pyplot as plt
import time 

def y(func, range):
    return [calculate_time_ms(func, i) for i in range]
        
def calculate_time_ms(func, n, show=False):
    start_time = time.process_time()
    process_value = func(n)
    if show:
        print(f"Value → None with 'n' arg as → {n}")
    elapsed = time.process_time() - start_time
    if show:
        print(f"Elapsed → {round(elapsed * 10**6, 2)} [µs], {round(elapsed * 10**3, 2)} [ms]\n")
    return elapsed * 10 ** 3

class DataHandeler:
    
    def plot(max):
        x = np.arange(0,max)
        plt.plot(x, y(np_array, x), "g")
        plt.plot(x, y(simple_array, x), "r")
        plt.title("Timing processing from different Array creation alghoritms")
        plt.legend(["In-build", "Iteration"])
        plt.xlabel('Index (n)')
        plt.ylabel('Elapsed (ms)')
        plt.show()
        
    def print_data_info(n):
        print("[Numpy array]:")
        calculate_time_ms(np_array, n, show=True)
        print("[Simple Array]:")
        calculate_time_ms(simple_array, n, show=True)  
```

```python
n = 2 * 10**3
DataHandeler.plot(n)
DataHandeler.print_data_info(2 * 10**6)
```

**Esercizio 28**   


Scrivi una funzione border(m,n,visualizza=False) che:
-      crea un array 2D di shape(m,n) che abbia tutti valori nulli
-     imposta ad 1  i valori degli elementi sul bordo dell'array (prima ed ultima riga, prima e l'ultima colonna)
-      Preveda un argomento di default che se impostato a True visualizza la matrice

```python
def border(size, visualizza=False):
    m = size[0]
    n = size[1]
    mat = np.zeros(size)
    mat[:,n-1:n] = 1
    mat[m-1:m,:] = 1
    mat[0:1,:] = 1
    mat[:,0:1] = 1
    if visualizza:
        print(mat)
```

```python
border((7, 8), True)
```
